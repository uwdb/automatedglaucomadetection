```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


tr_fn_bm3='tr_bm_gl_nm_bm3.txt'
val_fn_bm3='val_bm_gl_nm_bm3.txt'
te_fn_bm3='test_bm_gl_nm_bm3.txt'
sus_fn_bm3='sus_bm_gl_nm_bm3.txt'
```

```python
sus=pd.read_table(sus_fn_bm3)
sus=sus[['Age','BMI','FVC','PEF','IOP','l']]
te=pd.read_table(te_fn_bm3)
tr=pd.read_table(tr_fn_bm3)
v=pd.read_table(val_fn_bm3)
_a=pd.concat((te,tr),axis=0)
_a=pd.concat((_a,v),axis=0)
_a.shape
_a=_a[['Age','BMI','FVC','PEF','IOP','l']]
```

```python
nm=_a[_a['l']==0]
gl=_a[_a['l']==1]
sus['l']=0.5
comb=pd.concat((gl,nm),axis=0)
comb=pd.concat((comb,sus),axis=0)
```

```python
sns.set_style('whitegrid')
colors=['forestgreen','deeppink','darkorange']
sns.catplot(x='l',y='Age',kind="box",data=comb,palette=sns.color_palette(colors))

plt.savefig('ptg_age.png')
```

```python
sns.set_style('whitegrid')
colors=['forestgreen','deeppink','darkorange']
_sub=comb[comb['BMI']<41.7]
_sub=_sub[_sub['BMI']!=0]
sns.catplot(x='l',y='BMI',kind="box",data=_sub,palette=sns.color_palette(colors))
plt.savefig('ptg_bmi.png')
```

```python
_sub=comb[comb['IOP']<33]
_sub=_sub[_sub['IOP']!=0]
sns.catplot(x='l',y='IOP',kind="box",data=_sub,palette=sns.color_palette(colors))
plt.savefig('ptg_IOP.png')
```

```python
c=comb[comb['FVC']!=0]
c=c[c['FVC']<5.86]
sns.catplot(x='l',y='FVC',kind="box",data=c,palette=sns.color_palette(colors))

plt.savefig('ptg_FVC.png')
```
